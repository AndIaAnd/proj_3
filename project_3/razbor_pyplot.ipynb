```python
import matplotlib.animation as mpla
import numpy as np
plt.style.use('seaborn')
T = np.linspace(0, 2*np.pi, 100)
S = np.sin(T)
line, = plt.plot(T, S)
def animate(i):
    line.set_ydata(np.sin(T+i/50))
anim = mpla.FuncAnimation(
plt.gcf(), animate, interval=5)
plt.show()
```
